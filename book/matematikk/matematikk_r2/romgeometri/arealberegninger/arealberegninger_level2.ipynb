


````{admonition} L칮sningsforslag 
:class: dropdown

F칮rst kan vi dele sideflaten $BCGF$ i to trekanter, $BCG$ og $BGF$. Arealet av sideflaten blir da summen av arealene til de to trekantene.

Vi kan spenne ut $\Delta BCG$ ved 친 bruke vektorene

$$
\overrightarrow{BC} = [0, 4, 0] - [4, 4, 0] = [-4, 0, 0] \quad \text{og} \quad \vec{BG} = [1, 3, 3] - [4, 4, 0] = [-3, -1, 3].
$$

Arealvektoren for grunnflaten til $\Delta BCG$ blir da

$$
\vec{G}_{\Delta BCG} = \frac{1}{2} \vec{BC} \times \vec{BG} = \frac{1}{2} \begin{vmatrix} \vec{e}_x & \vec{e}_y & \vec{e}_z \\ -4 & 0 & 0 \\ -3 & -1 & 3 \end{vmatrix} = \frac{1}{2} [0, 12, 4] = [0, 6, 2] = 2 [0, 3, 1]
$$

Arealet av $\Delta BCG$ blir da

$$
G_{\Delta BCG} = |\vec{G}_{\Delta BCG}| = 2\sqrt{0^2 + 3^2 + 1^2} = 2\sqrt{10}.
$$

Vi kan spenne ut $\Delta BGF$ ved 친 bruke vektorene $\vec{BG}$ (som vi allerede kjenner) og $\vec{BF}$ som er gitt ved

$$
\vec{BF} = [3, 3, 3] - [4, 4, 0] = [-1, -1, 3].
$$

Grunnflatevektoren til $\Delta BGF$ blir da

$$

\vec{G}_{\Delta BGF} = \frac{1}{2} \vec{BG} \times \vec{BF} = \frac{1}{2} \begin{vmatrix} \vec{e}_x & \vec{e}_y & \vec{e}_z \\ -3 & -1 & 3 \\ -1 & -1 & 3 \end{vmatrix} = \frac{1}{2} [0, 6, 2] = [0, 3, 1].
$$

Arealet av $\Delta BGF$ blir da

$$
G_{\Delta BGF} = |\vec{G}_{\Delta BGF}| = \sqrt{0^2 + 3^2 + 1^2} = \sqrt{10}.
$$

Arealet av sidenflaten $BCGF$ blir da

$$
G_{BCGF} = G_{\Delta BCG} + G_{\Delta BGF} = 2\sqrt{10} + \sqrt{10} = 3\sqrt{10}.
$$
````

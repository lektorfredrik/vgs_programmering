
Eleven har startet p친 et program for 친 bruke ideen sin:

### Oppgave 3a 游닇
**Svar p친 f칮lgende sp칮rsm친l uten 친 kj칮re programmet:**
1. Forklar hvordan eleven har tenkt 친 bestemme en tiln칝rming til $\pi$. 
2. Hva tror du skjer n친r vi bruker veldig mange trekanter?




````{admonition} L칮sningsforslag
:class: dropdown

1. Eleven har tenkt at man kan tiln칝rme lengden av halvsirkelen ved 친 bruke lengden av rette linjer som f칮lger halvsirkelen som blir hypotenusen til trekanter som er laget mellom punktene p친 sirkelen og $x$-aksen. 
2. N친r vi bruker veldig mange trekanter, vil hypotenusene til trekantene ligge tiln칝rmet langs sirkelbuen. Summerer vi opp lengden av hypotenusene til trekantene, vil vi komme n칝rme lengden av sirkelbuen.

````


````{admonition} Mattehint: Kvadratrot som potens
:class: tip, dropdown

Kvadratroten $\sqrt{a}$ kan skrives som $a^{1/2}$. 
````


### Oppgave 3b 游눻

1. Regn ut en tiln칝rming for h친nd ved 친 bruke fire trekanter for 친 sjekke hva svaret omtrent b칮r bli. 游닇
2. Fullf칮r programmet til eleven slik at det regner ut en tiln칝rming til $\pi$ ved 친 bruke like mange trekanter som eleven opprinnelige hadde pr칮vd ut. 游눹 <br>


````{admonition} L칮sningsforslag
:class: dropdown

1. Vi kan regne ut en tiln칝rming for h친nd ved 친 bruke fire trekanter. Hvis vi deler intervallet $[-1, 1]$ i fire like store deler, f친r vi at $x$-verdiene til punktene blir

    $$
    x \in \left\{-1, -\frac{1}{2}, 0, \frac{1}{2}, 1\right\}.
    $$

    Funksjonsverdiene til 칮vre halvsirkel beskrevet av funksjonen 

    $$
    f(x) = \sqrt{1 - x^2}
    $$

    blir da

    $$
    f(x) \in \left\{0, \frac{\sqrt{3}}{2}, 1, \frac{\sqrt{3}}{2}, 0\right\}
    $$

    La $\Delta x$ ($x$-katet) v칝re avstanden mellom to n칝rliggende $x$-verdier og $\Delta y$ ($y$-katet) v칝re avstanden mellom to n칝rliggende $y$-verdier. Da blir lengden av hypotenusene $h$ til trekantene (etter Pytagoras' setning)

    $$
    h = \sqrt{(\Delta x)^2 + (\Delta y)^2}
    $$

    Vi kan lage oss en tabell der vi regner katetene og hypotenusene til trekantene: 

    | $\Delta x$ | $\Delta y$ | $h$ |
    |------------|------------|-----|
    |    $1/2$     | $\sqrt{3}/2$ | $\sqrt{\left(1/2\right)^2 + \left(\sqrt{3}/2\right)^2} = \sqrt{1/4 + 3/4} = 1$ |
    |    $1/2$     | $1 - \sqrt{3}/2$ | $\sqrt{\left(1/2\right)^2 + (1 - \sqrt{3}/2)^2} = \sqrt{1/4 + 1 - \sqrt{3} + 3/4} = \sqrt{2-\sqrt{3}}$ |
    |    $1/2$     | $\sqrt{3}/2 - 1$ | $\sqrt{\left(1/2\right)^2 + \left(\sqrt{3}/2 - 1\right)^2} = \sqrt{1/4 + 3/4 - \sqrt{3} + 1} = \sqrt{2-\sqrt{3}}$ |
    |    $1/2$     | $-\sqrt{3}/2$ | $\sqrt{\left(1/2\right)^2 + \left(-\sqrt{3}/2\right)^2} = \sqrt{1/4 + 3/4} = 1$ |


    

    Lengden av hypotenusene til trekantene blir derfor

    $$
    L = 2\cdot 1 + 2\cdot \sqrt{2 - \sqrt{3}} \approx 3.035.  
    $$



2. Programmet mangler 친 regne ut $y$-katetene, hypotenusen og summen av hypotenusene. Vi kan fullf칮re programmet slik:
    ```python
    def f(x):
        return (1 - x**2) ** 0.5

    a = -1
    b = 1
    x_katet = (b - a) / 4                                  # katet i x-retning i hver trekant med 4 trekanter

    x = a
    lengde = 0
    for i in range(4):
        # M친 regne ut y-katet i trekanten og hypotenus i trekanten
        y1 = f(x)                                           # y-verdi i trekantens venstre hj칮rne 
        y2 = f(x + x_katet)                                 # y-verdi i 칮vre hj칮rne
        y_katet = y2 - y1                                   # y-katet i trekanten (h칮yden)

        hypotenus = (x_katet**2 + y_katet**2)**0.5          # Hypotenus i trekanten
        lengde = lengde + hypotenus                         # Legger til hypotenus i lengde


        x = x + x_katet                                     # Flytter x til neste trekant 


    print(lengde)
    ```
    som gir utskriften
    ```console
    3.035276180410083
    ```
    som stemmer godt overens med regning vi har gjort for h친nd. 

````

### Oppgave 3c 游눹

Bruk programmet ditt til 친 finne en god tiln칝rming til $\pi$. 

$\pi$ er en irrasjonalt tall, s친 vi kan aldri f친 en eksakt verdi, men her er verdien med 11 desimaler:

$$
\pi \approx 3.14159265359
$$


Hvor mange trekanter trenger du for 친 f친 en tiln칝rming som har 11 desimaler riktig? 

````{admonition} Viktig kodehint: Antall trekanter m친 velges til 친 v칝re et partall!
:class: tip

P친 datamaskin, representeres tall med 2-tallssystemet, s친 for 친 unng친 avrundingsfeil m친 du velge antall trekanter til v칝re et partall (hvis ikke risikerer du 친 ta roten av et negativt tall p친 den ene enden). <br>

Dette kan du gj칮re ved sette for eksempel sette:

```python
n_trekanter = int(2**10)
```
men du m친 ha en h칮yere potens av $2$ for 친 f친 en god tiln칝rming til $\pi$ med 11 desimaler riktig. <br>
Merk ogs친 at programmet ditt vil ta lenger tid 친 kj칮re jo flere trekanter du bruker. Du kan forvente 친 m친tte vente rundt 10 sekunder n친r du kj칮rer programmet ved h칮yt nok antall trekanter.
````



````{admonition} L칮sningsforslag
:class: dropdown

Vi m친 칮ke antall trekanter i programmet for 친 finne et god tiln칝rming til lengden av buen til halvsirkelen.

```python
def f(x):
    return (1 - x**2) ** 0.5


n_trekanter = int(2**25)
a = -1
b = 1
x_katet = (b - a) / n_trekanter                         # katet i x-retning i hver trekant med n trekanter

x = a
lengde = 0
for i in range(n_trekanter):
    # M친 regne ut y-katet i trekanten og hypotenus i trekanten
    y1 = f(x)                                           # y-verdi i trekantens venstre hj칮rne 
    y2 = f(x + x_katet)                                 # y-verdi i 칮vre hj칮rne
    y_katet = y2 - y1                                   # y-katet i trekanten (h칮yden)

    hypotenus = (x_katet**2 + y_katet**2)**0.5          # Hypotenus i trekanten
    lengde = lengde + hypotenus                         # Legger til hypotenus i lengde


    x = x + x_katet                                     # Flytter x til neste trekant 

print(f"{lengde = :.11f}")
```
Ved 친 velge antall trekanter til 친 v칝re $2^{25} = 33554432$ f친r vi utskriften

```console
lengde = 3.14159265359
```
som gir oss en tiln칝rming til $\pi$ med 11 desimaler riktig. 
````

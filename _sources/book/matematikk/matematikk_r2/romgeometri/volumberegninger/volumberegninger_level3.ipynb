````{admonition} L칮sningsforslag
:class: dropdown

F칮rst og fremst m친 vi merke oss at grunnflaten *ikke* er et parallellogram, men et trapes. Derfor b칮r vi dele opp grunnflaten i to trekanter for 친 kunne bruke vektorregning. <br>
Vi kan for eksempel dele opp trapeset $ABCD$ i trekantene $\Delta ABC$ og $\Delta ACD$. <br>

Volumet av en pyramide med arealvektor $\vec{G}$ og "h칮ydevektor" $\vec{h}$ er gitt ved

$$
V = \frac{1}{3}|\vec{G}\cdot\vec{h}|.
$$

Pyramiden vi har her, vil best친 av to pyramiden med trekantet grunnflater, den ene med grunnflate $\Delta ABC$ og den andre med grunnflate $\Delta ACD$. <br>
Vi kan derfor finne volumet av pyramiden ved 친 summere volumene av de to pyramidene. <br>

Vi kan finne arealvektoren til $\Delta ABC$ ved 친 ta kryssproduktet av to vektorene som spenner ut trekantene:

$$
\overrightarrow{AB} = [2, 0, 0] - [0, 1, 0] = [2, -1, 0] \quad \text{og} \quad \overrightarrow{AC} = [2, 4, 0] - [0, 1, 0] = [2, 3, 0].
$$

Arealvektoren til $\Delta ABC$ er 

$$
\vec{G}_{\Delta ABC} = \frac{1}{2} \cdot \overrightarrow{AB} \times \overrightarrow{AC}.
$$

Vi kan finne kryssproduktet av de to vektorene ved

\begin{equation*}
\begin{split}
    \overrightarrow{AB} \times \overrightarrow{AC} & = 
    \begin{vmatrix} \vec{e}_x & \vec{e}_y & \vec{e}_z \\ 2 & -1 & 0 \\ 2 & 3 & 0 \end{vmatrix} \\ 
    \\
    & = \vec{e}_x \cdot ((-1) \cdot 0 - 0 \cdot 3) - \vec{e}_y \cdot (2 \cdot 0 - 0 \cdot 2) + \vec{e}_z \cdot (2 \cdot 3 - (-1) \cdot 2) \\
    \\
    & = \vec{e}_x \cdot 0 - \vec{e}_y \cdot 0 + \vec{e}_z \cdot 8 = [0, 0, 8] = 8 \cdot [0, 0, 1].
\end{split}
\end{equation*}

Vi trenger ogs친 arealvektoren til grunnflaten $\Delta ACD$. Vi kan finne denne p친 samme m친te som vi fant arealvektoren til $\Delta ABC$. <br>

F칮rst lager vi vektorene som spenner ut trekanten:

$$
\overrightarrow{AD} = [0, 3, 0] - [0, 1, 0] = [0, 2, 0] \quad \text{og} \quad \overrightarrow{AC} = [2, 4, 0] - [0, 1, 0] = [2, 3, 0].
$$

Deretter trenger vi kryssproduktet av de to vektorene:
\begin{equation*}
\begin{split}
    \overrightarrow{AD} \times \overrightarrow{AC} & = 
    \begin{vmatrix} \vec{e}_x & \vec{e}_y & \vec{e}_z \\ 0 & 2 & 0 \\ 2 & 3 & 0 \end{vmatrix} \\ 
    \\
    & = \vec{e}_x \cdot (2 \cdot 0 - 0 \cdot 3) - \vec{e}_y \cdot (0 \cdot 0 - 0 \cdot 2) + \vec{e}_z \cdot (0 \cdot 3 - 2 \cdot 2) \\
    \\
    & = \vec{e}_x \cdot 0 - \vec{e}_y \cdot 0 + \vec{e}_z \cdot 4 = [0, 0, 4] = 4 \cdot [0, 0, 1].
\end{split}
\end{equation*}

N친 kan vi bruke at volumet av de to pyramidene m친 gi volumet av hele pyramiden. Vi har at

$$
\vec{G}_{\Delta ABC} = \frac{1}{2}\cdot \overrightarrow{AB} \times \overrightarrow{AC} = \frac{1}{2} \cdot 8 \cdot [0, 0, 1] = 4 \cdot [0, 0, 1]
$$

og 

$$
\vec{G}_{\Delta ACD} = \frac{1}{2}\cdot \overrightarrow{AD} \times \overrightarrow{AC} = \frac{1}{2} \cdot 4 \cdot [0, 0, 1] = 2 \cdot [0, 0, 1].
$$

Begge grunnflater har et felles punkt i $A(0, 1, 0)$, s친 vi kan bruke dette punktet som toppunktet til pyramiden. <br> 
Vi kan lage denne ved 친 lage en vektor fra $A$ til $E$:

$$
\vec{h} = \overrightarrow{AE} = [0, 0, 3] - [0, 1, 0] = [0, -1, 3].
$$

Volumet av pyramiden (som er summen av volumene til de to pyramidene) er derfor
\begin{equation*}
\begin{split}
V & = \frac{1}{3}\cdot |\vec{G}_{\Delta ABC} \cdot \vec{h}| + \frac{1}{3}\cdot |\vec{G}_{\Delta ACD} \cdot \vec{h}| \\
\\
& = \frac{1}{3}\cdot |4 \cdot [0, 0, 1] \cdot [0, -1, 3]| + \frac{1}{3}\cdot |2 \cdot [0, 0, 1] \cdot [0, -1, 3]| \\
\\
& = \frac{1}{3}\cdot 4 \cdot \left(1\cdot 3\right) + \frac{1}{3}\cdot 2 \cdot \left(1\cdot 3\right) \\ 
\\
& = \frac{1}{3}\cdot 12 + \frac{1}{3}\cdot 6 = 4 + 2 \\
\\
& = 6.
\end{split}
\end{equation*}
````

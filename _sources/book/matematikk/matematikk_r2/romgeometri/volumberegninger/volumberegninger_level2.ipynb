


````{admonition} L칮sningsforslag
:class: dropdown

Volumet er av pyramide er generelt sett gitt ved

$$
V = \frac{1}{3} \cdot |\vec{G}\cdot \vec{h}|,
$$

der $\vec{G}$ er arealvektoren til grunnflaten og $\vec{h}$ er en vektor som g친r fra et av punktene i grunnflaten opp til toppunktet $T$ - vi kan tenke p친 det som "h칮ydevektoren" selv om det egentlig bare er der for 친 hjelpe oss 친 hente ut h칮yden ved hjelp av prikkproduktet. 

Grunnflaten til pyramiden er en trekant, som vi kan spenne ut ved hjelp av to vektorer. For eksempel kan vi velge oss $\overrightarrow{AB}$ og $\overrightarrow{AC}$ som to vektorer som spenner ut grunnflaten. Arealvektoren blir da

$$
\vec{G} = \frac{1}{2}\cdot \overrightarrow{AB} \times \overrightarrow{AC},
$$

der vi ganger med en faktor $1/2$ fordi det er en trekant. Vi kan regne ut kryssproduktet f칮rst:

\begin{equation*}
\begin{split}
\overrightarrow{AB} \times \overrightarrow{AC} & = \begin{vmatrix} \vec{e}_x & \vec{e}_y & \vec{e}_z \\ 5 & 0 & 0 \\ 4 & 2 & 0 \end{vmatrix} \\
\\
& = \vec{e}_x \cdot (0\cdot 0 - 2\cdot 0) - \vec{e}_y \cdot (5\cdot 0 - 4\cdot 0) + \vec{e}_z \cdot (5\cdot 2 - 4\cdot 0) \\
\\
&  = 10\vec{e}_z = [0, 0, 10] = 10\cdot [0, 0, 1].
\end{split}
\end{equation*}

Arealvektoren til grunnflaten blir da

$$
\vec{G} = \frac{1}{2}\cdot 10\cdot [0, 0, 1] = 5[0, 0, 1].
$$

Videre kan vi finne "h칮ydevektoren" ved 친 velge oss et vilk친rlig punkt i grunnflaten opp til toppunktet. Det enkleste er 친 velge $A(0, 0, 0)$ som utgangspunkt. H칮ydevektoren blir da

$$
\vec{h} = \overrightarrow{AT} = [0, 0, 5] - [0, 0, 0] = [0, 0, 5]
$$

som betyr at volumet blir

$$
V = \frac{1}{3}\cdot |\vec{G} \cdot \vec{h}| = \frac{1}{3} \cdot |5[0, 0, 1] \cdot [0, 0, 5]| = \frac{1}{3} \cdot \left|\left(5 \cdot \left(0\cdot 0 + 0 \cdot 0 + 1\cdot 5\right)\right)\right| = \frac{1}{3} \cdot 25 = \frac{25}{3}.
$$
````
